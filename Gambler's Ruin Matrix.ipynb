```python
import numpy as np
import scipy.linalg as la
import matplotlib.pyplot as plt
from numba import njit
```

```python
@njit(parallel = True)
def gamblers(n,p):
    R = np.eye(n-1)
    R[0,1] = -p
    R[n-2,n-3] = p-1
    for i in range(1,n-2):
        R[i,i-1] = p-1
        R[i,i+1] = -p
    V = np.zeros(n-1)
    V[n-2] = p
    
    pvec = np.linalg.inv(R)@V
    
    return pvec
```

```python
gr = gamblers(1000,0.49)
print(gr)
```

```python
sum(gr>=0.5)
```

```python
plt.figure(figsize = (15,12))
plt.semilogy(range(1,len(gr)+1),gr)
plt.semilogy(range(1,len(gr)+1),np.repeat(0.5,len(gr)),color = "k")
plt.scatter(range(1,len(gr)+1),gr)
plt.show()
```
